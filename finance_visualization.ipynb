```python
import pandas as pd

# Create the dataset
data = {
    "Date": ["2024-01-01","2024-02-01","2024-03-01",
             "2024-04-01","2024-05-01","2024-06-01"],
    "Revenue": [50000, 52000, 55000, 58000, 60000, 63000],
    "Expenses": [30000, 31000, 33000, 35000, 36000, 38000]
}

df = pd.DataFrame(data)
df

```

```python
# Convert Date to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Calculate Profit
df['Profit'] = df['Revenue'] - df['Expenses']

df
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.plot(df['Date'], df['Revenue'], marker='o')
plt.title("Revenue Over Time")
plt.xlabel("Date")
plt.ylabel("Revenue")
plt.show()
```

```python
plt.figure()
plt.bar(df['Date'].dt.month, df['Profit'])
plt.title("Monthly Profit")
plt.xlabel("Month")
plt.ylabel("Profit")
plt.show()
```

```python
import seaborn as sns

plt.figure()
sns.heatmap(df[['Revenue','Expenses','Profit']].corr(), annot=True)
plt.title("Financial Metrics Correlation")
plt.show()
```
